```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
from japanmap import picture
```

```python
df = pd.read_html("https://uub.jp/pdr/m/c.html")[4]
df = df.set_index(1)
```

```python
df.columns = df.loc["都道府県"]
df
```

```python
df.drop(["＃","合計↓"],axis=1,inplace=True)
```

```python
df.drop(["2020年3月末","都道府県","全国","▲"],inplace=True)
```

```python
df = df.replace("-",0).astype(int)
```

```python
df = df.idxmax(axis=1)
```

```python
cmap = {"セブンイレブン" : "yellow" ,"ファミリーマート" : "green" ,"ローソン" : "blue","セイコーマート":"red"}

fig = plt.subplots(figsize=(10,10))
plt.imshow(picture(df.replace(cmap)))
```
